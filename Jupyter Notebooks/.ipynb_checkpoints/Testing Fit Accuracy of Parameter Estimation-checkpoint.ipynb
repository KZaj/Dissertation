# Notebook for Evaluating the Performance of each of the OUP Parameter Estimation Methods

The methods described in the literature offer two different approaches to estimating the OUP parameters. To ensure we select the best one for use in our trading strategy, we evaluate the performance of each and make our selection. Deeper analysis of the results found in this notebook is included in the project report. 

```python
''' Module imports and declarations '''
%matplotlib inline
import math
import statistics
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.ticker as mticker
from scipy.stats import norm
from matplotlib.pyplot import figure
```

```python
''' Generate a batch/set of OUP paths, taken from the MATLAB example code '''
# alpha = mean reversion rate, mu = long-term mean, sigma = volatility
def gen_OUP_batch(alpha=5, mu=0.07, sigma=0.07, X0=0.03, T=1, nsteps=200, npaths=20000, analytic=True, plot=True, seed=-1):

    ''' Simulate an Ornstein-Uhlenbeck process '''
    # dX = alpha*(mu-X)*dt + sigma*dW
    
    # Seed for consistency if set
    if seed != -1:
        np.random.seed(seed)
    nsteps -= 1

    # Define parameters and time grid
    dt = T / nsteps                  # time step  
    t = np.linspace(0, T, nsteps+1)  # observation times

    ''' Allocate and initialise all paths '''
    X = [[X0] * npaths]
    for i in range(nsteps):
        X.append([0] * npaths)
    X = np.array(X)

    # Sample standard Gaussian random numbers
    N = np.random.normal(0, 1, (nsteps, npaths))

    ''' The following block is responsible for the actual generation of the OU paths '''
    # Euler with analytic moments
    if analytic:    
        
        # Compute the standard deviation for a time step
        sdev = sigma * math.sqrt((1 - math.exp(-2 * alpha * dt)) / (2 * alpha))  

        # Compute and accumulate the increments
        for i in range(nsteps):
            X[i + 1] = mu + (X[i] - mu) * math.exp(-alpha * dt) + sdev * N[i]
    
    # plain Euler 
    else:
        sdev = sigma * math.sqrt(dt)                                                                    
        for i in range(nsteps):
            X[i + 1] = X[i] + alpha * (mu - X[i]) * dt + sdev * N[i]
            

    # Compute the expected path
    EX = []
    for i in range(len(t)):
        val = mu + (X0 - mu) * math.exp(-alpha * t[i])
        EX.append(val)

    # Compute the mean path - Python alternative
    mean = []
    for i in range(len(X)):
        mean.append(np.mean(X[i]))

    if plot:
        ''' Plot the expected/mean/sample paths and mu '''
        figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')

        # This is from the original script, believe the first 3 lines are repeated at the end too
        # plot(t,EX,'k',t,mean(X,2),'k:',t,mu*ones(size(t)),'k--',t,X(:,1:1000:end),t,EX,'k',t,mean(X,2),'k:',t,mu*ones(size(t)),'k--')

        plt.plot(t, EX, 'k', label='Expected Path')
        plt.plot(t, mean, 'k:', label='Mean Path')
        plt.plot(t, [mu] * len(t), 'k--', label='mu')

        for i in range(10):
            plt.plot(t, X[:,i], linewidth=0.7)

        plt.legend()
        plt.xlabel('t')
        plt.ylabel('X')
        plt.title(r'Paths of an Ornstein-Uhlenbeck process $dX = \alpha(\mu-X)dt + \sigma dW$')
        # sdevinfty = sigma / math.sqrt(2 * alpha)
        # plt.ylim((mu-4*sdevinfty,mu+4*sdevinfty))
        plt.show()
    return X
```

# Least-squares Regression

The first of the two methods we evaluate is the least-squares regression approach. The methodology behind this approach is described in more detail in the project report. For each of the methods, we evaluate its ability to identify the parameters of OUP paths generated with known parameters. 

```python
''' Define a function that fits OUP using linear (least-squares regression) '''
def lin_fit_OUP(price_history, T=1, steps=200):
    
    # Estimate the parameters using the methods used in the literature
    n = len(price_history) - 1
    np_stock = np.array(price_history)
    
    Sx = sum(price_history[:-1])
    Sy = sum(price_history[1:])
    Sxx = np.sum(np_stock[:-1] ** 2)
    Sxy = np.sum(np_stock[:-1] * np_stock[1:])
    Syy = np.sum(np_stock[1:] ** 2)

    a = ((n * Sxy) - (Sx * Sy)) / ((n* Sxx) - (Sx ** 2))
    b = (Sy - (a * Sx)) / n
    sde = math.sqrt(((n * Syy) - (Sy ** 2) - (a * ((n * Sxy) - (Sx * Sy)))) / (n * (n - 2)))

    # dX = alpha*(mu-X)*dt + sigma*dW
    delta = T / steps

    # alpha = - ln a / delta
    alpha = - (math.log(a) / delta)

    # mu = b / 1 - a
    mu = b / (1 - a)

    # sigma = sd(epsilon) * sqrt( -2 ln a / delta * (1 - a^2) )
    sigma = sde * math.sqrt((-2 * math.log(a)) / (delta * (1 - a ** 2)))  
    return alpha, mu, sigma
```

```python
''' Parameters to discover: alpha, mu and sigma '''
# Parameters that fitting and generation both need access to
t = 1
stp = 1000

# Testing the fitting accuracy with 20000 paths:
paths = 20000
a = 5
m = 0.07
s = 0.03
series = gen_OUP_batch(alpha=a, mu=m, sigma=s, T=t, nsteps=stp, npaths=paths, plot=False, seed=1)

# For plotting
alphas = []
mus = []
sigmas = []

# For calculating the average error
a_ = []
m_ = []
s_ = []

# Keeping track of averages
a_tot, s_tot, m_tot = 0, 0, 0

# Extract the parameters for each series and average them out
for i in range(paths):
    alpha, mu, sigma = lin_fit_OUP(series[:,i], T=t, steps=stp)
    a_tot += alpha
    m_tot += mu
    s_tot += sigma
    
    a_.append(alpha)
    m_.append(mu)
    s_.append(sigma)
    
    alphas.append(a_tot / (i + 1))
    mus.append(m_tot / (i + 1))
    sigmas.append(s_tot / (i + 1))
    
# Plot the results of the fitting    
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(alphas, label="alpha")
plt.ylabel("Lambda")
plt.xlabel("Number of series sampled")
plt.title("Lambda estimation with each sampled series")
plt.show()

figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(mus, label="mu")
plt.ylabel("Mu")
plt.xlabel("Number of series sampled")
plt.title("Mu estimation with each sampled series")
plt.show()

figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(sigmas, label="sigma")
plt.ylabel("Sigma")
plt.xlabel("Number of series sampled")
plt.title("Sigma estimation with each sampled series")
plt.show()

print("Alpha: ", alphas[-1], " | Mu: ", mus[-1], " | Sigma: ", sigmas[-1])
```

# Maximum Likelihood Estimators

In the following blocks, we repeat the same analysis as we do for the least-squares regression, but now with this second method. Again, this method is described in more detail in the project report. 

```python
''' Define a function that fits OUP using Maximum Likelihood Estimators '''
def max_fit_OUP(price_history, T=1, steps=200):
    
    n = len(price_history) - 1
    np_stock = np.array(price_history)

    Sx = sum(price_history[:-1])
    Sy = sum(price_history[1:])

    Sxx = np.sum(np_stock[:-1] ** 2)
    Sxy = np.sum(np_stock[:-1] * np_stock[1:])
    Syy = np.sum(np_stock[1:] ** 2)

    # dX = alpha*(mu-X)*dt + sigma*dW
    delta = T / steps

    # MU
    mu = ((Sy * Sxx) - (Sx * Sxy))/ ((n * (Sxx - Sxy)) - ((Sx ** 2) - (Sx * Sy)))
    
    # LAMBDA (ALPHA)
    delta = T / steps
    alpha = - (1 / delta) * math.log((Sxy - (mu * Sx) - (mu * Sy) + (n * (mu ** 2))) / (Sxx - (2 * mu * Sx) + (n * (mu **2))))
    
    # SIGMA
    a = math.exp(-(alpha * delta))
    sigma_dot = (1 / n) * (Syy - (2 * a * Sxy) + ((a ** 2) * Sxx) - \
                           (2 * mu * (1 - a) * (Sy - (a * Sx))) + (n * (mu ** 2) * ((1 - a)**2)))
    sigma = math.sqrt((sigma_dot) * ((2 * alpha) / (1 - (a ** 2))))
    return alpha, mu, sigma
```

```python
''' COPIED BLOCK AS FOR THE LEAST-SQUARES, BUT NOW DONE WITH MLE '''
# Parameters that fitting and generation both need access to
t = 1
stp = 1000

# Testing the fitting accuracy with 20000 paths:
paths = 20000
a = 5
m = 0.07
s = 0.03
series = gen_OUP_batch(alpha=a, mu=m, sigma=s, T=t, nsteps=stp, npaths=paths, plot=False, seed=1)

# For plotting
alphas = []
mus = []
sigmas = []

# For calculating the average error
a_ = []
m_ = []
s_ = []

# Keeping track of averages
a_tot, s_tot, m_tot = 0, 0, 0

# Extract the parameters for each series and average them out
for i in range(paths):
    alpha, mu, sigma = max_fit_OUP(series[:,i], T=1, steps=stp)
    a_tot += alpha
    m_tot += mu
    s_tot += sigma
    
    a_.append(alpha)
    m_.append(mu)
    s_.append(sigma)
    
    alphas.append(a_tot / (i + 1))
    mus.append(m_tot / (i + 1))
    sigmas.append(s_tot / (i + 1))
    
# Plot the results of the fitting    
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(alphas, label="alpha")
plt.ylabel("Lambda")
plt.xlabel("Number of series sampled")
plt.title("Lambda estimation with each sampled series")
plt.show()

figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(mus, label="mu")
plt.ylabel("Mu")
plt.xlabel("Number of series sampled")
plt.title("Mu estimation with each sampled series")
plt.show()

figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(sigmas, label="sigma")
plt.ylabel("Sigma")
plt.xlabel("Number of series sampled")
plt.title("Sigma estimation with each sampled series")
plt.show()

print("Alpha: ", alphas[-1], " | Mu: ", mus[-1], " | Sigma: ", sigmas[-1])
```

Performing a further investigation into why we see the incorrect results for the estimation of lambda, reveals that the use of the time horizon in the fitting plays a larger role than anticipated. While performing a further analysis of this is outside of our scope, we investigate these effects to better showcase the accuracy of our estimation methods. The results are further analysed in the project report. 

```python
''' Test the effects of changing the value of T against the accuracy of estimation '''
# Parameters that fitting and generation both need access to
stp = 1000
Ts = [3, 10, 100]

# Testing the fitting accuracy with 20000 paths:
paths = 20000
a = 5
m = 0.07
s = 0.03

alpha_lines = []
mu_lines = []
sigma_lines = []

for t in Ts:
    series = gen_OUP_batch(alpha=a, mu=m, sigma=s, T=t, nsteps=stp, npaths=paths, plot=False, seed=1)

    # For plotting
    alphas = []
    mus = []
    sigmas = []

    # For calculating the average error
    a_ = []
    m_ = []
    s_ = []

    # Keeping track of averages
    a_tot, s_tot, m_tot = 0, 0, 0

    # Extract the parameters for each series and average them out
    for i in range(paths):
        
        ''' FITTING AND GENERATION HAVE ACCESS TO THE SAME T '''
        alpha, mu, sigma = lin_fit_OUP(series[:,i], T=t, steps=stp)
        a_tot += alpha
        m_tot += mu
        s_tot += sigma

        a_.append(alpha)
        m_.append(mu)
        s_.append(sigma)

        alphas.append(a_tot / (i + 1))
        mus.append(m_tot / (i + 1))
        sigmas.append(s_tot / (i + 1))
    
    alpha_lines.append(alphas)
    mu_lines.append(mus)
    sigma_lines.append(sigmas)
    
# Plot the results of the fitting    
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
for i in range(len(Ts)):
    plt.plot(alpha_lines[i], label="T: " + str(Ts[i]))
plt.plot([a] * paths, '--',label="Actual lambda")
plt.ylabel("Lambda")
plt.xlabel("Number of series sampled")
plt.title("Lambda estimation with each sampled series")
plt.legend()
plt.show()

figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
for i in range(len(Ts)):
    plt.plot(mu_lines[i], label="T: " + str(Ts[i]))
plt.plot([m] * paths, '--',label="Actual mu")
plt.ylabel("Mu")
plt.xlabel("Number of series sampled")
plt.title("Mu estimation with each sampled series")
plt.legend()
plt.show()

figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
# plt.plot(sigmas, label="sigma")
for i in range(len(Ts)):
    plt.plot(sigma_lines[i], label="T: " + str(Ts[i]))
plt.plot([s] * paths, '--',label="Actual sigma")
plt.ylabel("Sigma")
plt.xlabel("Number of series sampled")
plt.title("Sigma estimation with each sampled series")
plt.legend()
plt.show()

for i in range(len(Ts)):
    print(i, "Alpha: ", alpha_lines[i][-1], " | Mu: ", mu_lines[i][-1], " | Sigma: ", sigma_lines[i][-1])
```

# Investigating the Effects of Fitting Reduced Sample

We also investigate the effects of reducing the sample size the fitting has access to. Since we apply our estimation methods in this way, i.e. we use only a part of the full series at each observation, we must investigate these effects to better understand how our strategy behaves. 

```python
# Parameters that fitting and generation both need access to
stp = 1000
t=5

# Testing the fitting accuracy across a few paths
paths = 5
a = 5
m = 0.07
s = 0.03
series = gen_OUP_batch(alpha=a, mu=m, sigma=s, T=t, nsteps=stp, npaths=paths, plot=False, seed=4)

# For sampling the series
minimum_sample = 5
maximum_sample = 1000

all_a = []
all_m = []
all_s = []

# Extract the parameters for each series and plot them accordingly
for i in range(paths):
    alpha_history = []
    mu_history = []
    sigma_history = []

    for point in range(minimum_sample, maximum_sample):
        
        # The fitting might fail due to negative alpha value, in which case, just ignore this setting
        # This happens rarely enough that the differences are not visible on the graphs at this scale
        try:
            alpha, mu, sigma = lin_fit_OUP(series[:,i][0:point], T=t, steps=stp)
            alpha_history.append(alpha)
            mu_history.append(mu)
            sigma_history.append(sigma)
        except:
            continue

    all_a.append(alpha_history)
    all_m.append(mu_history)
    all_s.append(sigma_history)
    
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
for i in range(paths):
    plt.plot(range(minimum_sample, len(all_a[i]) + minimum_sample), all_a[i], label="OUP path " + str(i))
plt.plot([a]*maximum_sample, '--', label="Expected")
plt.legend()
plt.xlabel("Length of series observed")
plt.ylabel("Estimated value of lambda")
plt.title("Value estimation of lambda with reduced sample of series")
# plt.ylim(1.5, 12)
plt.show()

figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
for i in range(paths):
    plt.plot(range(minimum_sample, len(all_m[i]) + minimum_sample), all_m[i], label="OUP path " + str(i))
plt.plot([m]*maximum_sample, '--', label="Expected")
plt.legend()
plt.xlabel("Length of series observed")
plt.ylabel("Estimated value of mu")
plt.title("Value estimation of mu with reduced sample of series")
# plt.ylim(-0.05, 0.2)
plt.show()

figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
for i in range(paths):
    plt.plot(range(minimum_sample, len(all_s[i]) + minimum_sample), all_s[i], label="OUP path " + str(i))
plt.plot([s]*maximum_sample, '--', label="Expected")
plt.legend()
plt.xlabel("Length of series observed")
plt.ylabel("Estimated value of sigma")
plt.title("Value estimation of sigma with reduced sample of series")
# plt.ylim(0.022, 0.035)
plt.show()
```

# Investigating Cases of Negative Values of a when Estimating

The value of a plays a large role in the estimation, in particular, two of the OUP parameters rely on a natural logarithm of the value of a. Since this is undefined when the value of a is negative, we seek to investigate when these negatives occur so we can effectively deal with them in the context of our trading strategy.

```python
''' Reduced function from the estimation method, for investigating the value of a '''
def a_verbose(price_history):
    n = len(price_history) - 1
    np_stock = np.array(price_history)
    Sx = sum(price_history[:-1])
    Sy = sum(price_history[1:])
    Sxx = np.sum(np_stock[:-1] ** 2)
    Sxy = np.sum(np_stock[:-1] * np_stock[1:])
    a = ((n * Sxy) - (Sx * Sy)) / ((n * Sxx) - (Sx ** 2))
    return a

# Parameters that fitting and generation both need access to
stp = 1000
t=1

# Test the frequency of negative a values across 2000 paths
paths = 2000
a = 5
m = 0.07
s = 0.03
series = gen_OUP_batch(alpha=a, mu=m, sigma=s, T=t, nsteps=stp, npaths=paths, plot=False, seed=4)

# This corresponds to the maximum length of the sample used for fitting
l = 100
counts = [0] * l

# Keep track of negatives at each sample size
for n in range(paths):
    a_s = []
    for point in range(0, l):
        a = a_verbose(series[:,n][0:point+1])
        if a < 0:
            counts[point] += 1

# Plot the results
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(counts[:50])
plt.xlabel("Size of Series Sampled")
plt.ylabel("No. of Negative 'a' Calculated (out of 2000)")
plt.title("Sample Size Effect on Fitting - Parameter 'a'")
plt.show()
```
