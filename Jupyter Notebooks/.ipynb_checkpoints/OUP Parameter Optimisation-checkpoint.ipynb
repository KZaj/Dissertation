# Notebook for Parameter Optimisation for the OUP Strategy

```python
''' Module imports and declarations '''
%matplotlib inline
import math
import statistics
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.ticker as mticker
from scipy.stats import norm
from matplotlib.pyplot import figure
```

```python
''' Define a function that fits OUP using linear (least-squares regression) '''
def lin_fit_OUP(price_history, T=1, steps=200):
    
    # Estimate the parameters using the methods used in the literature
    n = len(price_history) - 1
    np_stock = np.array(price_history)
    
    Sx = sum(price_history[:-1])
    Sy = sum(price_history[1:])
    Sxx = np.sum(np_stock[:-1] ** 2)
    Sxy = np.sum(np_stock[:-1] * np_stock[1:])
    Syy = np.sum(np_stock[1:] ** 2)

    a = ((n * Sxy) - (Sx * Sy)) / ((n* Sxx) - (Sx ** 2))
    b = (Sy - (a * Sx)) / n
    sde = math.sqrt(((n * Syy) - (Sy ** 2) - (a * ((n * Sxy) - (Sx * Sy)))) / (n * (n - 2)))

    # dX = alpha*(mu-X)*dt + sigma*dW
    delta = T / steps

    # alpha = - ln a / delta
    alpha = - (math.log(a) / delta)

    # mu = b / 1 - a
    mu = b / (1 - a)

    # sigma = sd(epsilon) * sqrt( -2 ln a / delta * (1 - a^2) )
    sigma = sde * math.sqrt((-2 * math.log(a)) / (delta * (1 - a ** 2)))  
    return alpha, mu, sigma
```

```python
''' 
DISCLAIMER: this code is largely taken from the Numerical Methods for Finance UCL Moodle Page,
          specifically, the OUP MATLAB code as suggested by the project supervisor 
'''

''' Function for generating OUP series '''
def gen_OUP(alpha=5, mu=0.07, sigma=0.07, seed=-1, init_val=0.03, T=1, steps=200, analytic=True):
    
    ################################################
    # alpha, mu and sigma are the model parameters #
    # dX = alpha*(mu-X)*dt + sigma*dW              #
    #                                              #
    # alpha is the mean-reversion strength         #
    # mu is the value of the long-term mean        # 
    # sigma is the volatility of the series        #
    ################################################
    
    # If the seed is set, use it, otherwise generate randomly
    if seed != -1:
        np.random.seed(seed)

    # Store the series
    X = [init_val]
    
    # The initial value is considered the "first step"
    steps -= 1
    
    # Sample standard Gaussian random numbers
    mean = 0
    std  = 1
    N = np.random.normal(mean, std, steps)

    # Calulate the time difference with each step
    dt = T / steps

    ''' The following block is responsible for the actual generation of the OU paths '''
    # Euler with analytic moments
    if analytic:
        # Compute the standard deviation for a time step
        sdev = sigma * math.sqrt((1 - math.exp(-2 * alpha * dt)) / (2 * alpha))
        
        # Compute and accumulate the increments
        for i in range(steps):
            X.append(mu + (X[i] - mu) * math.exp(-alpha * dt) + sdev * N[i])
            
    # plain Euler 
    else:
        sdev = sigma * math.sqrt(dt)                                               
        for i in range(steps):
            X.append(X[i] + alpha * (mu * X[i]) * dt + sdev * N[i])
            
    # Return the generated series
    return X
```

```python
''' Trade OUP with selected parameters - back-test '''
def trade_OUP(stock_price, trading_time=500, ma_period=21, ma_t="ema", T=1, quantile=0.25, plot=False):
    
    # Ensure the parameters are valid
    s_len = len(stock_price)
    assert trading_time + ma_period < s_len, "Trading time with moving average period must be smaller than series length."

    # Keep track of lines for plotting
    top_band = []
    bot_band = []
    ma = []
    
    # Keep track of accumulated profit
    profit_percent = 0
    
    # Current state of trading
    holding = False
    buy_point = None
    
    # If the moving average is ema, we need an initial value
    if ma_t == "ema":
        ema = stock_price[s_len - trading_time - ma_period]
        for index in range(s_len - trading_time - ma_period + 1, s_len - trading_time):
            ema = (stock_price[index] - ema) * (2 / (ma_period + 1)) + ema
        ma.append(ema)
    
    # Go through the series and trade OUP
    for point_index in range(s_len - trading_time, s_len):
        
        # Identify the indexes for the start and end of the relevant series
        start = point_index - ma_period + 1
        end = point_index + 1
        
        # Exponential moving average
        if ma_t == "ema":
            ema = (stock_price[point_index] - ma[-1]) * (2 / (ma_period + 1)) + ma[-1]
            ma.append(ema)
        
        # Simple moving average 
        else:
            sma = sum(stock_price[start:end]) / ma_period
            ma.append(sma) 
        
        # Define the rest of the corridor
        try:
            
            # We are looking for the probability distribution at point 'point_index', using the series 
            # of len ma_period before it
            alpha, mu, sigma = lin_fit_OUP(stock_price[start:end], steps=ma_period, T=T)

            # Calculate the probability distribution at this point
            expected = mu + ((stock_price[start] - mu) * math.exp(-alpha * T))
            standard_dev = sigma * math.sqrt((1 - math.exp(-2 * alpha * T)) / (2 * alpha))
            quantiles = norm.ppf([quantile, 0.5, 1 - quantile], expected, standard_dev)
            
            # Calculate the offsets
            top = ma[-1] + (quantiles[2] - quantiles[1])
            bot = ma[-1] - (quantiles[1] - quantiles[0])

        # In the event we cannot perform the fit with least-squares
        except:
            
            # Use the previous offset if we can
            try:
                offset_top = top_band[-1] - ma[-2]
                top = ma[-1] + offset_top
                
                offset_bot = ma[-2] - bot_band[-1]
                bot = ma[-1] - offset_bot
                
            # Otherwise, use offset = 5% of price
            except:
                offset = 0.05 * stock_price[point_index]
                top = ma[-1] + offset
                bot = ma[-1] - offset
        
        # Append the top and bottom regardless of how we generate them
        top_band.append(top)
        bot_band.append(bot)
        
        # Figure out if we need to buy or sell
        if stock_price[point_index] <= bot and not holding:
            holding = True
            buy_point = point_index   
        elif stock_price[point_index] >= top and holding:
            holding = False
            
            # If we sell, accumulate the profit as a percentage
            profit_percent += ((stock_price[point_index] - stock_price[buy_point]) / (stock_price[buy_point])) * 100
            
    # Plot everything relevant to trading
    if plot:
        figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')        
        plt.plot(stock_price, label="Asset")
        plt.plot(range(s_len-trading_time, s_len), top_band, label="Top Band")
        plt.plot(range(s_len-trading_time, s_len), bot_band, label="Bottom Band")
        
        # Handle the differences between ema and sma generation
        if ma_t == "ema":
            plt.plot(range(s_len-trading_time, s_len), ma[1:], label="EMA")
        else:
            plt.plot(range(s_len-trading_time, s_len), ma, label="SMA")
        plt.ylim(min(stock_price), max(stock_price))
        plt.xlabel("Observations")
        plt.ylabel("Asset Value")
        plt.legend()
    return profit_percent
```

# Moving Average Type

Short evaluation of the importance of the moving average type, we compare the performance across a series of 500 assets, with varying parameters. The parameters have to be manually adjusted for the ones the user wants to test. This block largely just serves for the population of the tables we use in the project report. 

```python
# Generate a set of OUP paths
assets = []
for i in range(500):
    asset = gen_OUP(alpha=1, mu=100, init_val=100, sigma=10, steps=1000)
    assets.append(asset)

# Keep track of which one "wins"
s = 0
d = 0
e = 0

# Keep track of totals for average profitability
s_tot = 0
e_tot = 0

# Parameters to test
MA = 300
QUANT = 0.4

# For each asset, attempt both simple and exponential moving averages
for i in range(len(assets)):
    sm = trade_OUP(assets[i], ma_t="sma", ma_period=MA, quantile=QUANT)
    em = trade_OUP(assets[i], ma_t="ema", ma_period=MA, quantile=QUANT) 
    print("SMA: ", sm, " \t| EMA: ", em)

    # Increase the totals depending on the results
    s_tot += sm
    e_tot += em
    
    if sm > em:
        s += 1
    elif em > sm:
        e += 1
    else:
        d += 1
    
# Print basic formatting results to the console
print("\n" + str(s) + " " + str(e) + " " + str(d))
print(s_tot/len(assets), e_tot/len(assets))
```

# Moving Average Period

Here we test the direct effects of the moving average period on the profitability. The code is pretty self explanatory, we seek to vary the period and record the effects on profits. 

```python
# Generate the assets to test against and prepare the graph for plotting
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k') 
for i in range(5):
    assets = []
    for _ in range(500):
        asset = gen_OUP(alpha=1, mu=100, init_val=100, sigma=10, steps=1000)
        assets.append(asset)
        
    # We test across the following moving average period values
    MAS = [7, 21, 50, 100, 150, 200, 250, 300, 350, 400]
    line = []
    for MA in MAS:
        print("MA: ", MA)
        profits = []
        
        # Take the average profits across the assets in this set
        for asset in assets:
            profit = trade_OUP(asset, ma_period=MA, quantile=0.3)
            profits.append(profit)
            print(profit)
        avg = sum(profits)/len(profits)
        line.append(avg)
    plt.plot(MAS, line, label="Series set " + str(i+1))
    
# Plot results and add labels
plt.xlabel("Moving Average Period")
plt.ylabel("Profit/Loss (%)")
plt.legend()
plt.title("Moving average period effect on profitability")
plt.show()
```

# Quantile

We repeat the same sorts of optimisations as for the moving average period, but this time with the sampled quantile. The code blocks are essentially identical. We break this code up into two different blocks, to display the effects of averaging across several moving average periods in this analysis.

```python
# Keep track of all of the results
all_data = []
for i in range(3):
    
    # Generate a set of OUP paths
    assets = []
    for i in range(500):
        asset = gen_OUP(alpha=1, mu=100, init_val=100, sigma=10, steps=1000)
        assets.append(asset)

    # Prepare a plot
    figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')  

    # Iterate over these parameters
    MAS = [30, 100, 200, 300]
    QUANTS = [0.05, 0.1, 0.15, 0.2, 0.25, 0.3, \
              0.35, 0.4, 0.45, 0.5]
    all_quant_profs = []
    for MA in MAS:
        print("Period: ", MA)
        quant_prof = []
        for QUANT in QUANTS:
            print("\tQuant: ", QUANT),
            profits = []
            
            # Average performance across the assets
            for i in range(len(assets)):
                profit = trade_OUP(assets[i], ma_period=MA, quantile=QUANT)
                profits.append(profit)
            avg_prof = sum(profits)/len(profits)
            print("\t", avg_prof)
            quant_prof.append(avg_prof)

        # Plot intermediate results but continue to append to all_data
        plt.plot(QUANTS, quant_prof, label=str(MA))
        all_quant_profs.append(quant_prof)
    all_data.append(all_quant_profs)

    # Plot intermediate
    plt.legend(title="Moving Average Period")
    plt.ylabel("Profit/Loss (%)")
    plt.xlabel("Quantile")
    plt.title("Effect of Quantile Selection on Profitability")
    plt.show()
```

```python
# Plot the results of averaging the above
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k') 
for i, all_quant_profs in enumerate(all_data):    
    np_profs = np.array(all_quant_profs)
    line = []
    for q in range(len(QUANTS)):
        column = np_profs[:,q]
        print(sum(column), column)
        line.append(sum(column)/len(column))
    plt.plot(QUANTS, line, label="Series set " + str(i))
plt.legend()
plt.title("Effects of quantile selection on profitability")
plt.xlabel("Quantile")
plt.ylabel("Profit/Loss (%)")
plt.show()
```

# Combined Optimisation

In these following blocks we aim to investigate the effects of both of the parameters being altered in conjunction. Since the parameters may have residual effects on one another, it does not suffice to investigate their effect independently. 

```python
# Generate assets to test agains
assets = []
for i in range(500):
    asset = gen_OUP(alpha=1, mu=100, init_val=100, sigma=10, steps=1000)
    assets.append(asset)
    
# Optimise across these parameter combinations
MAS = [25, 50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325, 350, 375, 400]
QUANTS = [0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45]

# Keep track of results
all_combos = []
for ma in MAS:
    print("MA: ", ma)
    long_combos = []
    for q in QUANTS:
        print("Quant: ", q)
        profits = []
        
        # Average the profits across all of the combinations
        for asset in assets:
            prof = trade_OUP(asset, ma_period=ma, quantile=q)
            profits.append(prof)
        avg = sum(profits) / len(profits)
        print(avg)
        long_combos.append(round(avg, 3))
    long_combos = long_combos
    all_combos.append(long_combos)
print(np.array(all_combos))
```

```python
# Convert the results into colour values 
abs_min = 100
abs_max = -100
for row in all_combos:
    for elem in row:
        if elem < abs_min:
            abs_min = elem
        if elem > abs_max:
            abs_max = elem
abs_max += abs(abs_min)        
colour_vals = []
for row in all_combos:
    new_row = []
    for elem in row:
        val = 1 - ((elem + abs(abs_min)) / abs_max)
        new_row.append(round(val, 3))
    colour_vals.append(new_row)
print(np.array(colour_vals))

# Convert results into an easy to plot heatmap
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k') 
xs = []
ys = []
cols = []
for i, ma in enumerate(MAS):
    for j, q in enumerate(QUANTS):
        xs.append(ma)
        ys.append(q)
        cols.append(colour_vals[i][j])
plt.scatter(xs, ys, c=cols, s=300, cmap='inferno')
plt.xlabel('Moving Average Period')
plt.ylabel('Sampled Quantile')
plt.title('Parameter effects on profitability of OUP strategy')
plt.show()
```

# OUP with Commission

Following the investigation into the effects of each parameter on the trading frequency, we explore how an added commission impacts the profitability of the strategy. We add a 5% commission to each trade made, where a 5% charge means you pay 5% of the asset price in order to begin the trade.

```python
''' Trade OUP with selected parameters - modified to include commission '''
def trade_OUP_com(stock_price, trading_time=500, ma_period=21, ma_t="ema", T=1, quantile=0.25, plot=False):
    
    # Ensure the parameters are valid
    s_len = len(stock_price)
    assert trading_time + ma_period < s_len, "Trading time with moving average period must be smaller than series length."

    # Keep track of lines for plotting
    top_band = []
    bot_band = []
    ma = []
    
    # Keep track of accumulated profit
    profit_percent = 0
    
    # Current state of trading
    holding = False
    buy_point = None
    
    # If the moving average is ema, we need an initial value
    if ma_t == "ema":
        ema = stock_price[s_len - trading_time - ma_period]
        for index in range(s_len - trading_time - ma_period + 1, s_len - trading_time):
            ema = (stock_price[index] - ema) * (2 / (ma_period + 1)) + ema
        ma.append(ema)
    
    # Go through the series and trade OUP
    for point_index in range(s_len - trading_time, s_len):
        
        # Identify the indexes for the start and end of the relevant series
        start = point_index - ma_period + 1
        end = point_index + 1
        
        # Exponential moving average
        if ma_t == "ema":
            ema = (stock_price[point_index] - ma[-1]) * (2 / (ma_period + 1)) + ma[-1]
            ma.append(ema)
        
        # Simple moving average 
        else:
            sma = sum(stock_price[start:end]) / ma_period
            ma.append(sma) 
        
        # Define the rest of the corridor
        try:
            
            # We are looking for the probability distribution at point 'point_index', using the series 
            # of len ma_period before it
            alpha, mu, sigma = lin_fit_OUP(stock_price[start:end], steps=ma_period, T=T)

            # Calculate the probability distribution at this point
            expected = mu + ((stock_price[start] - mu) * math.exp(-alpha * T))
            standard_dev = sigma * math.sqrt((1 - math.exp(-2 * alpha * T)) / (2 * alpha))
            quantiles = norm.ppf([quantile, 0.5, 1 - quantile], expected, standard_dev)
            
            # Calculate the offsets
            top = ma[-1] + (quantiles[2] - quantiles[1])
            bot = ma[-1] - (quantiles[1] - quantiles[0])

        # In the event we cannot perform the fit with least-squares
        except:
            
            # Use the previous offset if we can
            try:
                offset_top = top_band[-1] - ma[-2]
                top = ma[-1] + offset_top
                
                offset_bot = ma[-2] - bot_band[-1]
                bot = ma[-1] - offset_bot
                
            # Otherwise, use offset = 5% of price
            except:
                offset = 0.05 * stock_price[point_index]
                top = ma[-1] + offset
                bot = ma[-1] - offset
        
        # Append the top and bottom regardless of how we generate them
        top_band.append(top)
        bot_band.append(bot)
        
        # Figure out if we need to buy or sell
        if stock_price[point_index] <= bot and not holding:
            holding = True
            buy_point = point_index   
        elif stock_price[point_index] >= top and holding:
            holding = False
            
            # If we sell, accumulate the profit as a percentage
            profit = ((stock_price[point_index] - stock_price[buy_point]) / (stock_price[buy_point])) * 100
            profit -= (stock_price[buy_point]) * 0.05
            profit_percent += profit
            
    # Plot everything relevant to trading
    if plot:
        figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')        
        plt.plot(stock_price, label="Asset")
        plt.plot(range(s_len-trading_time, s_len), top_band, label="Top Band")
        plt.plot(range(s_len-trading_time, s_len), bot_band, label="Bottom Band")
        
        # Handle the differences between ema and sma generation
        if ma_t == "ema":
            plt.plot(range(s_len-trading_time, s_len), ma[1:], label="EMA")
        else:
            plt.plot(range(s_len-trading_time, s_len), ma, label="SMA")
        plt.ylim(min(stock_price), max(stock_price))
        plt.xlabel("Observations")
        plt.ylabel("Asset Value")
        plt.legend()
    return profit_percent
```

```python
# Perform the same sort of performance evaluation as we carry out earlier in this notebook
t=1
assets = []
for i in range(500):
    asset = gen_OUP(alpha=1, mu=100, init_val=100, sigma=10, steps=1000)
    assets.append(asset)
MAS = [25, 50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325, 350, 375, 400]
QUANTS = [0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45]
all_combos = []
for ma in MAS:
    print("MA: ", ma)
    long_combos = []
    for q in QUANTS:
        print("Quant: ", q)
        profits = []
        for asset in assets:
            prof = trade_OUP_com(asset, ma_period=ma, quantile=q)
            profits.append(prof)
        avg = sum(profits) / len(profits)
        print(avg)
        long_combos.append(round(avg, 3))
    long_combos = long_combos
    all_combos.append(long_combos)
print(np.array(all_combos))
```
