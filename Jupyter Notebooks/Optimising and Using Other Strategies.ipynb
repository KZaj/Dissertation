```python
''' Module imports and declarations '''
%matplotlib inline
import math
import statistics
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.ticker as mticker
from scipy.stats import norm
from matplotlib.pyplot import figure
```

```python
''' 
DISCLAIMER: this code is largely taken from the Numerical Methods for Finance UCL Moodle Page,
          specifically, the OUP MATLAB code as suggested by the project supervisor 
'''

''' Function for generating OUP series '''
def gen_OUP(alpha=5, mu=0.07, sigma=0.07, seed=-1, init_val=0.03, T=1, steps=200, analytic=True):
    
    ################################################
    # alpha, mu and sigma are the model parameters #
    # dX = alpha*(mu-X)*dt + sigma*dW              #
    #                                              #
    # alpha is the mean-reversion strength         #
    # mu is the value of the long-term mean        # 
    # sigma is the volatility of the series        #
    ################################################
    
    # If the seed is set, use it, otherwise generate randomly
    if seed != -1:
        np.random.seed(seed)

    # Store the series
    X = [init_val]
    
    # The initial value is considered the "first step"
    steps -= 1
    
    # Sample standard Gaussian random numbers
    mean = 0
    std  = 1
    N = np.random.normal(mean, std, steps)

    # Calulate the time difference with each step
    dt = T / steps

    ''' The following block is responsible for the actual generation of the OU paths '''
    # Euler with analytic moments
    if analytic:
        # Compute the standard deviation for a time step
        sdev = sigma * math.sqrt((1 - math.exp(-2 * alpha * dt)) / (2 * alpha))
        
        # Compute and accumulate the increments
        for i in range(steps):
            X.append(mu + (X[i] - mu) * math.exp(-alpha * dt) + sdev * N[i])
            
    # plain Euler 
    else:
        sdev = sigma * math.sqrt(dt)                                               
        for i in range(steps):
            X.append(X[i] + alpha * (mu * X[i]) * dt + sdev * N[i])
            
    # Return the generated series
    return X
```

# Moving Average Crossovers

The moving average crossover strategy is detailed in the contents of the project report. Essentially, the strategy relies on trading on a trend-following basis. When the short moving average is above the long moving average, buy, otherwise, sell. The thought process here is to catch the majority of the profit on an upward shift, and more or less ignore the asset otherwise.

```python
''' Trade Moving Average Crossovers - back-test '''
def trade_MA(stock_price, trading_time=500, short_p=7, long_p=21, plot=False):
    
    # Ensure the parameters are valid
    assert short_p < long_p, "Short period must be shorter than long period."
    assert trading_time + long_p < len(stock_price), "Moving average history and trading time must be less than total length."

    # Keep track of lines for plotting
    ma_s = []
    ma_l = []
    
    # Keep track of accumulated profit
    profit_percent = 0
    
    # Current state of trading
    holding = False
    buy_point = None

    # Go through the series and trade crossovers
    for point_index in range(len(stock_price) - trading_time, len(stock_price)):

        # Calculate short MA and append for plot
        short = sum(stock_price[point_index - short_p + 1:point_index + 1]) / short_p
        ma_s.append(short)
        
        # Repeat for long MA
        long = sum(stock_price[point_index - long_p + 1:point_index + 1]) / long_p
        ma_l.append(long)
        
        # Figure out if we need to buy or sell
        if short > long and not holding:
            holding = True
            buy_point = point_index
        elif short < long and holding:
            holding = False
            
            # If we sell, accumulate the profit made as a percentage
            profit_percent += ((stock_price[point_index] - stock_price[buy_point]) / (stock_price[buy_point])) * 100
            
    # Plot everything relevant to trading
    if plot:    
        figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k') 
        plt.plot(stock_price, label="Asset")
        plt.plot(range(len(stock_price)-trading_time, len(stock_price)), ma_s, label="Short MA")
        plt.plot(range(len(stock_price)-trading_time, len(stock_price)), ma_l, label="Long MA")
        plt.xlabel("Observations")
        plt.ylabel("Asset Value")
        plt.ylim(min(stock_price), max(stock_price))
        plt.legend()
    return profit_percent

# Test the strategy on a randomly generated OUP series
asset = gen_OUP(alpha=1, mu=100, init_val=100, sigma=10, steps=1000)
trade_MA(asset, plot=True)
```

```python
''' Produce a plot of the most optimal parameters for the moving average strategy '''
# Generate a set of 500 OUP paths with a stable long-term mean
assets = []
for i in range(500):
    asset = gen_OUP(alpha=1, mu=100, init_val=100, sigma=10, steps=1000)
    assets.append(asset)
    
# Optimise across the following moving average period values
MAS = [25, 50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325, 350, 375, 400]

# Keep track of the performance of each combo
all_combos = []
for i, long in enumerate(MAS[1:]):
    
    # Print to console to track progress
    print("Long: ", long)
    
    # Performance of each short MA for this long MA
    long_combos = []
    for short in MAS[:i+1]:
        
        # Console print
        print("Short: ", short)
        
        # Calculate the average profit across the 500 paths for this combination
        profits = []
        for asset in assets:
            prof = trade_MA(asset, short_p=short, long_p=long)
            profits.append(prof)
        avg = sum(profits) / len(profits)
        print(avg)
        long_combos.append(round(avg, 3))
        
    # Conver to np.array for simplicity later
    long_combos = np.array(long_combos)
    all_combos.append(long_combos)
print(np.array(all_combos))
```

```python
''' Go through the results array and make some alterations to make a heatmap plot '''
# Identify the minimum and maximum values
arr = np.array(all_combos)
abs_min = 100
abs_max = -100
for row in arr:
    for elem in row:
        if elem < abs_min:
            abs_min = elem
        if elem > abs_max:
            abs_max = elem
print("Max:", abs_max, "| Min:", abs_min)
abs_max += abs(abs_min)

# The colour corresponds to profitability, and must satisfy 0 <= value <= 1
colour_vals = []
for row in arr:
    new_row = []
    for elem in row:
        
        # 1 - (...) makes it so that darker colours correspond to more profit rather than lighter
        val = 1 - ((elem + abs(abs_min)) / abs_max)
        new_row.append(round(val, 3))
    colour_vals.append(new_row)
print(colour_vals)

# Conver the results to a set of arrays that make this heatmap easy to plot
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
xs = []
ys = []
cols = []
for i, long in enumerate(MAS[1:]):
    for j, short in enumerate(MAS[:i+1]):
        xs.append(short)
        ys.append(long)
        cols.append(colour_vals[i][j])

# Generate the parameter optimisation plot
plt.scatter(xs, ys, c=cols, s=200, cmap='inferno')
plt.xlabel("Short MA Period")
plt.ylabel("Long MA Period")
plt.title("Profitability with varying short/long moving average periods")
plt.show()
```

# Bollinger Bands

```python
''' Trade Bollinger Bands - back-test '''
def trade_BB(stock_price, trading_time=500, ma_period=21, std_devs=1, plot=False):
    
    # Ensure the parameters are valid
    s_len = len(stock_price)
    assert trading_time + ma_period < s_len, "Trading time with moving average period must be smaller than series length."

    # Keep track of lines for plotting
    top_band = []
    bot_band = []
    ma = []
    
    # Keep track of accumulated profit
    profit_percent = 0
    
    # Current state of trading
    holding = False
    buy_point = None
            
    # Initialise the initial value of exponential moving average
    ema = stock_price[s_len - trading_time - ma_period]
    for index in range(s_len - trading_time - ma_period + 1, s_len - trading_time):
        ema = (stock_price[index] - ema) * (2 / (ma_period + 1)) + ema
    ma.append(ema)
    
    # Go through the series and trade OUP
    for point_index in range(s_len - trading_time, s_len):
        
        # Identify the indexes for the start and end of the relevant series
        start = point_index - ma_period + 1
        end = point_index + 1
        
        # Update EMA
        ema = (stock_price[point_index] - ma[-1]) * (2 / (ma_period + 1)) + ma[-1]
        ma.append(ema)

        # Get standard deviation
        sdev = np.std(stock_price[start:end])
        
        # Set the top and bottom bands with an offset equal to the standard deviation
        top = ema + (std_devs * sdev)
        bot = ema - (std_devs * sdev)
        
        # Append the top and bottom for plotting
        top_band.append(top)
        bot_band.append(bot)
        
        # Figure out if we need to buy or sell
        if stock_price[point_index] <= bot and not holding:
            holding = True
            buy_point = point_index
        elif stock_price[point_index] >= top and holding:
            holding = False
            
            # If we sell, accumulate the profit as a percentage
            profit_percent += ((stock_price[point_index] - stock_price[buy_point]) / (stock_price[buy_point])) * 100
            
    # Plot everything relevant to trading
    if plot:      
        figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
        plt.plot(stock_price, label="Asset")
        plt.plot(range(s_len-trading_time, s_len), top_band, label="Top Band")
        plt.plot(range(s_len-trading_time, s_len), bot_band, label="Bottom Band")
        plt.plot(range(s_len-trading_time, s_len), ma[1:], label="EMA")  
        plt.xlabel("Observations")
        plt.ylabel("Asset Value")
        plt.legend()
        plt.ylim(min(stock_price), max(stock_price))
    return profit_percent

# Test the strategy on a randomly generated OUP series
asset = gen_OUP(alpha=1, mu=100, init_val=100, sigma=10, steps=1000)
trade_BB(asset, ma_period=200, plot=True)
```

```python
# Perform the same conversion to colour values as we see before
abs_min = 100
abs_max = -100
for row in all_combos:
    for elem in row:
        if elem < abs_min:
            abs_min = elem
        if elem > abs_max:
            abs_max = elem    
abs_max += abs(abs_min)        
colour_vals = []
for row in all_combos:
    new_row = []
    for elem in row:
        val = 1 - ((elem + abs(abs_min)) / abs_max)
        new_row.append(round(val, 3))
    colour_vals.append(new_row)
print(np.array(colour_vals))

# Conver to easy to plot arrays
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k') 
xs = []
ys = []
cols = []
for i, ma in enumerate(MAS):
    for j, d in enumerate(DEV):
        xs.append(ma)
        ys.append(d)
        cols.append(colour_vals[i][j])
        
# Plot the optimisation graph
plt.scatter(xs, ys, c=cols, s=300, cmap='inferno')
plt.xlabel('Moving Average Period')
plt.ylabel('No. of Standard Deviations Offset')
plt.title('Parameter effects on profitability of bollinger band strategy')
plt.show()
```

# Strategy Comparisons on Real Data

The following analysis is used in the report during the bench-marking section. Here we seek to analyse the performance of each of the strategies on particular series that are either generated with a particular long-term trend, or are taken from historic data. 

```python
''' Define a function that fits OUP using linear (least-squares regression) '''
def lin_fit_OUP(price_history, T=1, steps=200):
    
    # Estimate the parameters using the methods used in the literature
    n = len(price_history) - 1
    np_stock = np.array(price_history)
    
    Sx = sum(price_history[:-1])
    Sy = sum(price_history[1:])
    Sxx = np.sum(np_stock[:-1] ** 2)
    Sxy = np.sum(np_stock[:-1] * np_stock[1:])
    Syy = np.sum(np_stock[1:] ** 2)

    a = ((n * Sxy) - (Sx * Sy)) / ((n* Sxx) - (Sx ** 2))
    b = (Sy - (a * Sx)) / n
    sde = math.sqrt(((n * Syy) - (Sy ** 2) - (a * ((n * Sxy) - (Sx * Sy)))) / (n * (n - 2)))

    # dX = alpha*(mu-X)*dt + sigma*dW
    delta = T / steps

    # alpha = - ln a / delta
    alpha = - (math.log(a) / delta)

    # mu = b / 1 - a
    mu = b / (1 - a)

    # sigma = sd(epsilon) * sqrt( -2 ln a / delta * (1 - a^2) )
    sigma = sde * math.sqrt((-2 * math.log(a)) / (delta * (1 - a ** 2)))  
    return alpha, mu, sigma
```

```python
''' Trade OUP with selected parameters - back-test '''
def trade_OUP(stock_price, trading_time=500, ma_period=21, ma_t="ema", T=1, quantile=0.25, plot=False):
    
    # Ensure the parameters are valid
    s_len = len(stock_price)
    assert trading_time + ma_period < s_len, "Trading time with moving average period must be smaller than series length."

    # Keep track of lines for plotting
    top_band = []
    bot_band = []
    ma = []
    
    # Keep track of accumulated profit
    profit_percent = 0
    
    # Current state of trading
    holding = False
    buy_point = None
    
    # If the moving average is ema, we need an initial value
    if ma_t == "ema":
        ema = stock_price[s_len - trading_time - ma_period]
        for index in range(s_len - trading_time - ma_period + 1, s_len - trading_time):
            ema = (stock_price[index] - ema) * (2 / (ma_period + 1)) + ema
        ma.append(ema)
    
    # Go through the series and trade OUP
    for point_index in range(s_len - trading_time, s_len):
        
        # Identify the indexes for the start and end of the relevant series
        start = point_index - ma_period + 1
        end = point_index + 1
        
        # Exponential moving average
        if ma_t == "ema":
            ema = (stock_price[point_index] - ma[-1]) * (2 / (ma_period + 1)) + ma[-1]
            ma.append(ema)
        
        # Simple moving average 
        else:
            sma = sum(stock_price[start:end]) / ma_period
            ma.append(sma) 
        
        # Define the rest of the corridor
        try:
            
            # We are looking for the probability distribution at point 'point_index', using the series 
            # of len ma_period before it
            alpha, mu, sigma = lin_fit_OUP(stock_price[start:end], steps=ma_period, T=T)

            # Calculate the probability distribution at this point
            expected = mu + ((stock_price[start] - mu) * math.exp(-alpha * T))
            standard_dev = sigma * math.sqrt((1 - math.exp(-2 * alpha * T)) / (2 * alpha))
            quantiles = norm.ppf([quantile, 0.5, 1 - quantile], expected, standard_dev)
            
            # Calculate the offsets
            top = ma[-1] + (quantiles[2] - quantiles[1])
            bot = ma[-1] - (quantiles[1] - quantiles[0])

        # In the event we cannot perform the fit with least-squares
        except:
            
            # Use the previous offset if we can
            try:
                offset_top = top_band[-1] - ma[-2]
                top = ma[-1] + offset_top
                
                offset_bot = ma[-2] - bot_band[-1]
                bot = ma[-1] - offset_bot
                
            # Otherwise, use offset = 5% of price
            except:
                offset = 0.05 * stock_price[point_index]
                top = ma[-1] + offset
                bot = ma[-1] - offset
        
        # Append the top and bottom regardless of how we generate them
        top_band.append(top)
        bot_band.append(bot)
        
        # Figure out if we need to buy or sell
        if stock_price[point_index] <= bot and not holding:
            holding = True
            buy_point = point_index   
        elif stock_price[point_index] >= top and holding:
            holding = False
            
            # If we sell, accumulate the profit as a percentage
            profit_percent += ((stock_price[point_index] - stock_price[buy_point]) / (stock_price[buy_point])) * 100
            
    # Plot everything relevant to trading
    if plot:
        figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')        
        plt.plot(stock_price, label="Asset")
        plt.plot(range(s_len-trading_time, s_len), top_band, label="Top Band")
        plt.plot(range(s_len-trading_time, s_len), bot_band, label="Bottom Band")
        
        # Handle the differences between ema and sma generation
        if ma_t == "ema":
            plt.plot(range(s_len-trading_time, s_len), ma[1:], label="EMA")
        else:
            plt.plot(range(s_len-trading_time, s_len), ma, label="SMA")
        plt.ylim(min(stock_price), max(stock_price))
        plt.xlabel("Observations")
        plt.ylabel("Asset Value")
        plt.legend()
    return profit_percent
```
